```python
# Common imports
import pandas as pd
import numpy as np

# For plot/visualizations
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

#Import functions from utils
from utils import plot_dataset
from utils import save_fig
from utils import fit_normal_equation

# ignoring Numpy error
np.seterr(divide = 'ignore')

np.random.seed(1)
```

```python
# load the preprocessed iris dataset
dataset = pd.read_csv('datasets/Advertising.csv', header=0, names=['TV', 'radio', 'newspaper', 'sales'])
print('\nNumber of records:',len(dataset))
print('\n\n',dataset.head())
```

```python
# Setting up training data
X = pd.DataFrame(dataset,columns=['TV']).values
y = dataset['sales'].values
print('The shape of X:',X.shape)
print('The shape of y:',y.shape)
```

```python
plt.figure(1)
plot_dataset(X,y)
plt.title('Figure 1: TV vs. Sales')
plt.axis([0.0,300,0.0,30.0])
plt.legend()
save_fig("dataset")
plt.show()
```

```python
plt.figure(2)
plt.title('Figure 2: Linear Model using Normal Equation')
plot_dataset(X,y)
fit_normal_equation(X,y)
plt.axis([0.0,300,0.0,30.0])
plt.legend(loc="lower right")
save_fig("fit_linear_equation")
plt.show()
```
