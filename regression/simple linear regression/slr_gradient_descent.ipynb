```python
# Common imports
import numpy as np
import os

# For plot/visualizations
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Imports from utils
from utils import save_fig, plot_dataset, fit_linear_regression
from utils import mse_function, hypothesis
from utils import normal_equation, gradient_descent

# ignoring Numpy error
np.seterr(divide = 'ignore')

np.random.seed(1)
```

```python
# Defining the training dataset
X = np.random.rand(200,1).reshape(-1,1)  
y = 3*X + .5 + np.random.normal(size=200).reshape(-1,1)
```

```python
plt.figure(1)
plot_dataset(X,y)
plt.title('Figure 1: Sample Training Datapoints')
plt.axis([-0.1,1.1,-3.0,6.0])
plt.legend()
#save_fig("dataset")
plt.show()
```

```python
# Initial settings
X = np.insert(X,0,1,axis=1)
(m,n) = X.shape 
init_theta = np.zeros((n,1)) 
iterations = 1500  
learning_rate = 1
```

```python
error = mse_function(X, y, init_theta);
print('Initial cost: %.2f'% error);
```

```python
theta = gradient_descent(X, y, init_theta, learning_rate, iterations);
print('Minimum point(found by gradient descent):', theta[:,0])
```

```python
plt.figure(2)
plt.title('Figure 2: Linear Model using Batch Gradient Descent')
plot_dataset(X[:,1],y)
fit_linear_regression(X,theta)
plt.axis([-0.1,1.1,-3.0,6.0])
plt.legend()
plt.show()
```

```python
theta_ne = normal_equation(X,y)[:,0]
print("\nMinimum point(found by Normal Equation):",theta)
```
