```python
# Common imports
import pandas as pd
import numpy as np

# For plot/visualizations
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# For optimization
import scipy.optimize as opt

# Import from utils
from utils import save_fig
from utils import plot_dataset
from utils import feature_standardise
from utils import mean_squared_error

# ignoring Numpy error
np.seterr(divide = 'ignore')

np.random.seed(0)

%load_ext autoreload
%autoreload 2
```

```python
# load the preprocessed iris dataset
dataset = pd.read_csv('datasets/Auto.csv', header=0)
print('\nNumber of records:',len(dataset))
print('\n\n',dataset.head())
```

```python
dataset = dataset.replace('?',np.NaN)

print('Number of instances = %d' % (dataset.shape[0]))
print('Number of attributes = %d' % (dataset.shape[1]))
print('Number of missing values:')
for col in dataset.columns:
    print('\t%s: %d' % (col,dataset[col].isna().sum()))
```

```python
print('Number of rows in original data = %d' % (dataset.shape[0]))
data = dataset.dropna()
print('Number of rows after discarding missing values = %d' % (data.shape[0]))
```

```python
# Setting up training and testdata
X = pd.to_numeric(pd.DataFrame(data)["horsepower"]).values.reshape(-1,1)
X = feature_standardise(X)
y = data['mpg'].values.reshape(-1,1)

# Hyperparameter, regularization term, Lambda value
Lambda = 0.00
```

```python
plt.figure(1)
plt.title('Figure 1: Horsepower vs. Miles per gallon')
plot_dataset(X,y)
plt.axis([-0.4,.8,5,50])
plt.legend()
save_fig("ann_datapoints")
plt.show()
```

```python
def layer_sizes(X,y):
    input_layer = X.shape[1]
    hidden_layer = 10
    output_layer = y.shape[1]
    return input_layer, hidden_layer, output_layer
```

```python
def initialize_parameters(layer_in,layer_out):
    random_epsilon = np.sqrt(6) / np.sqrt(layer_in+layer_out)
    theta = (np.random.random((layer_out, layer_in + 1))*
             (2*random_epsilon)) - random_epsilon
    return theta
```

```python
def feed_forward(X,theta):
    # Unroll the theta values
    W1 = theta[:(input_layer+1)*hidden_layer].reshape((hidden_layer,(input_layer+1)))
    W2 = theta[(input_layer+1)*hidden_layer:].reshape((output_layer,(hidden_layer+1)))
    
    # Forward propagation
    A1 = np.insert(X,0,1,axis=1)
    Z2 = A1.dot(W1.T)
    A2 = np.insert(np.tanh(Z2),0,1,axis=1) 
    Z3 = A2.dot(W2.T)

    cache = {"A1":A1,
             "Z2":Z2, 
             "A2":A2, 
             "Z3":Z3}
    return Z3, cache
```

```python
def backward_propagation(theta,cache,y,Lambda):
    # --------------------------
    # Retrieving parameters
    # --------------------------
    W1 = theta[:(input_layer+1)*hidden_layer].reshape((hidden_layer,(input_layer+1)))
    W2 = theta[(input_layer+1)*hidden_layer:].reshape((output_layer,(hidden_layer+1)))

    # -------------------------------------------------------------
    # Retrieving cached activations
    # -------------------------------------------------------------
    A1 = cache["A1"]
    Z2 = cache["Z2"]
    A2 = cache["A2"]
    Z3 = cache["Z3"]

    # -------------------------------------------------------------
    # The Backward Propagation Implementation
    # -------------------------------------------------------------
    d3 = (Z3 - y)
    u  = np.tanh(Z2)
    d2 = (Z3 - y).dot(W2[:,1:]) * (1 - np.power(u,2))
    D1 = d2.T.dot(A1) / len(y)
    D2 = d3.T.dot(A2) / len(y)
    if Lambda!=0:
        D1[:,1:] += (Lambda/len(y) * W1[:,1:])
        D2[:,1:] += (Lambda/len(y) * W2[:,1:])

     # Roll grads D1 and D2 back into 1D array
    grads = np.concatenate((D1.reshape(((input_layer+1)*hidden_layer)),
                            D2.reshape(((hidden_layer+1)*output_layer))))
    return grads
```

```python
def predict(X,theta):
    return feed_forward(X,theta)[0]
```

```python
# Setting network structure
input_layer,hidden_layer, output_layer = layer_sizes(X,y)
# Initialize and combine thetas
theta_init = np.concatenate((initialize_parameters(input_layer,hidden_layer).reshape(((input_layer+1)*hidden_layer)),
                             initialize_parameters(hidden_layer,output_layer).reshape(((hidden_layer+1)*output_layer))))
```

```python
def nn_model(theta_init,X,y,Lambda):
    # Forward propagation
    Z3, cache = feed_forward(X, theta_init)

    # cross-entropy loss function
    cost = mean_squared_error(Z3,y)
    
    # Backpropagation
    grads = backward_propagation(theta_init,cache,y,Lambda) 

    return cost, grads
```

```python
def nn_cost_function(theta,X,y,Lambda):
    return nn_model(theta,X,y,Lambda)[0]
```

```python
print("Initial cost:",nn_cost_function(theta_init,X,y,Lambda))
```

```python
def fit(theta_init,X,y,Lambda):
    obj_func = lambda t: nn_cost_function(t,X,y,Lambda);
    result = opt.minimize(obj_func, x0=theta_init, method='BFGS')
    return result.x, result.fun
```

```python
theta, error = fit(theta_init,X,y,Lambda);
print("Cost: %f" %(error))
print("Mean Squared Error(MSE): %f" %(mean_squared_error(y, predict(X,theta))))
```

```python
def plot_nn_fit(X,theta):
    Xvals = np.arange(X.min() - 25, X.max() + 25, 0.05).reshape(-1,1)
    y_pred = predict(Xvals,theta)
    plt.plot(Xvals,y_pred, 'r--',linewidth=2, label="ANN")
```

```python
plt.figure(2)
plot_dataset(X,y)
plot_nn_fit(X,theta)
plt.title('Figure 2: Linear Regression with ANN (Lambda = %.2f)' % (Lambda));
plt.axis([-0.4,0.8,5,50])
plt.legend(loc="upper right")
save_fig("lr_ann")
plt.show()
```
