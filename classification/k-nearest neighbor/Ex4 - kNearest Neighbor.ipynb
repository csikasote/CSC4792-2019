```python
# Import common python libraries
import numpy as np
import pandas as pd
import os

# Import function from utils
from utils import save_fig
from utils import plot_dataset
from utils import KNearestNeighbor
from utils import knn_decision

# Matplotlib for plotting figures
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# load the preprocessed iris dataset
dataset = pd.read_csv('datasets/marketing.csv', delimiter=',', header=0, names=["ID","Salary","Age","Purchase"])

print('\nNumber of records:',len(dataset))
print('\nclass domain:', dataset['Purchase'].unique())
print('\n\n',dataset.head())
```

```python
X = pd.DataFrame(dataset,columns=['Age','Salary']).values
y = dataset['Purchase'].values
print('The shape of X:',X.shape)
print('The shape of y:',y.shape)
```

```python
plt.figure(1)
plt.title('Fig. 1: KNN classifier: Age vs. Salary')
plot_dataset(X,y)
plt.xlabel("Age", fontsize=14)
plt.ylabel("Salary", fontsize=14)
plt.axis([25,65,40000,75000])
plt.grid(True)
save_fig("Ex4_Plot")
plt.show()
```

```python
knn = KNearestNeighbor(k=3)
knn.train(X, y)
Xtest = np.array([[35, 56000]])
distances = knn.distance(Xtest)
print(distances)
y_hat = knn.predict(Xtest)
print("\nKNN decision:",knn_decision(y_hat))
```

```python
plt.figure(2)
plt.title('Fig. 2: KNN classifier: Age vs. Salary')
plot_dataset(X,y)
plt.plot(Xtest[0,0], Xtest[0,1],'s', color='blue', markersize=10, label="Test Customer")
plt.xlabel("Age", fontsize=14)
plt.ylabel("Salary", fontsize=14)
plt.axis([25,65,40000,75000])
plt.grid(True)
plt.legend()
save_fig("kNN_TestExample")
plt.show()
```
