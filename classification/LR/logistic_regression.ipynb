```python
# Import common python libraries
import numpy as np
import pandas as pd
import os

# For optimization
from scipy import optimize as opt

# Matplotlib for plotting figures
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

# ignoring Numpy error
np.seterr(divide = 'ignore') 

# Where to save the figures
EXERCISE_ROOT_DIR = "."
IMAGES_PATH = os.path.join(EXERCISE_ROOT_DIR, "images")
```

```python
# The function allows images to be saved
def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
# load the preprocessed iris dataset
dataset = pd.read_csv('datasets/iris_process.data',
                      delimiter=',',
                      header=None, 
                      names=['sepal length', 'sepal width', 'petal length', 'petal width','class'])

print('\nNumber of records:',len(dataset))
print('\nclass domain:', dataset['class'].unique())
print('\n\n',dataset.head())
```

```python
X = pd.DataFrame(dataset,columns=['sepal length','sepal width']).values
y = dataset['class'].values
print('The shape of X:',X.shape)
print('The shape of y:',y.shape)
```

```python
def plot_dataset(X,y):
    plt.scatter(X[y == 0, 0],X[y == 0, 1],
                s=50, c='lightgreen',
                marker='o', edgecolor='black',label='class 1')
    plt.scatter(X[y == 1, 0],X[y == 1, 1],
                s=50, c='orange',
                marker='v', edgecolor='black',
                label='class 2')
    plt.tight_layout()
    plt.xlabel("sepal length", fontsize=14)
    plt.ylabel("sepal width", fontsize=14)
```

```python
plt.figure(1)
plt.title('Iris data: sepal length vs. sepal width')
plot_dataset(X,y)
plt.grid(True)
plt.axis([4.0,8.0,1.5,4.5])
plt.legend()
#save_fig("lr_training_data")
plt.show()
```

```python
def sigmoid(z):
    return (1/(1+ np.exp(-z)))
```

```python
x_vals = np.linspace(-10, 10, 50) # DEFINING THE X AXIS 
def sigmoidPlot(x):
    y = sigmoid(x)
    plt.plot(x,y, 'k--', label='sigmoid/logistic Function')
    plt.plot(0,0.5,'X')
    plt.ylabel(r"$\sigma(.)$", fontsize=17, rotation=0)
    plt.xlabel(r"$z$", fontsize=17)
    plt.xlim([-10,10])
    plt.ylim([0,1])
```

```python
plt.figure(2)
plt.title("sigmoid function")
sigmoidPlot(x_vals)
plt.grid(True)
plt.legend()
#save_fig("lr_sigmoid")
plt.show()
```

```python
# INITIALIZATIONS
(m,n) = X.shape
num_iters = 400
theta = np.zeros((n+1, 1)); # initialize fitting parameters to zeros
Xnew = np.c_[np.ones((m, 1)), X]; # Add a column of ones to x for x0 = 1
```

```python
def z(x, theta):
    return np.dot(x,theta)
```

```python
def cost_function(theta, x, y):
    h = z(x, theta)
    y_hat = sigmoid(h)
    cost = (-1/len(x)) * (np.dot(y.T,np.log(y_hat)) + np.dot((1-y).T,np.log(1-y_hat)))
    return cost 
```

```python
cost = cost_function(theta, Xnew, y) # Calling the logistic cost function
print('\nInitial Cost: %f' % (cost) + ', Theta: ' + str(theta[:,0]))
```

```python
def fit(func, theta,x,y,iters):
    result = opt.fmin(func=func, x0=theta, args=(x,y),maxiter=iters, full_output=1)
    return result[0], result[1]
```

```python
theta,cost = fit(cost_function,theta,Xnew,y, num_iters)
print('\nResults (found by fmin):\nMinimum point: '\
      + str(theta) + ', Cost: %.5f' % (cost));
```

```python
def predict(X,theta):
    probability = sigmoid(z(X,theta))
    if probability>=0.5:
        return "class 2"
    elif probability < 0.5:
        return "class 1"
```

```python
Xtest = [1, 6.75,4.25]
Ytest = predict(Xtest,theta)
print("\nPrediction:",Ytest)
```

```python
def plot_decision_boundary(X,y,theta):
    theta0 = theta[0]; 
    theta1 = theta[1]; 
    theta2 = theta[2];
    fig, ax = plt.subplots()
    plot_dataset(X,y)
    x_vals = np.array(ax.get_xlim())
    y_vals = -1 * np.divide(((np.multiply(theta1,x_vals)) + theta0),theta2)
    plt.plot(x_vals, y_vals, '--', c="red", label='Decision Boundary')
    titlestr = 'Decision Boundary: y = %.2f + %.2fX1 + %.2fX2' % (theta0, theta1, theta2)
    plt.title(titlestr)
    plt.grid(True)
    plt.legend()
```

```python
plt.figure(3)
plot_decision_boundary(X,y, theta)
plt.plot(6.75,4.25,'s', color='blue', markersize=10)
plt.annotate("X=(6.75, 4.25)",
             xy=(6.78, 4.25),
             xytext=(6.8, 4.4),
             ha="center",
             fontsize=8,
            )
plt.axis([4.0,8.0,1.5,4.5])
#save_fig("lr_decision")
plt.show()
```
